```python
import pandas as pd
from sqlalchemy import create_engine



# SQL-Query ausf端hren und in DataFrame laden
query = """
SELECT
    c.customer_id,
    c.signup_date,
    co.contract_id,
    co.start_date,
    fch.churn_reason,
    ft.revenue
FROM dw.dim_customer c
LEFT JOIN dw.dim_contract co
    ON c.customer_id = co.customer_id
LEFT JOIN dw.fact_transaction ft
    ON c.customer_id = ft.customer_id
LEFT JOIN dw.fact_churn fch
    ON c.customer_id = fch.customer_id
"""

df = pd.read_sql(query, engine)

# df zeigt jetzt alle relevanten Daten
df.head()
```

```python
df['churned'] = df['churn_reason'].notnull().astype(int)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sicherstellen, dass Datum als Datetime vorliegt
df['signup_date'] = pd.to_datetime(df['signup_date'])

# Cohort-Monat bestimmen
df['cohort_month'] = df['signup_date'].dt.to_period('M')

# Cohort-Matrix (Kunden pro Kohorte)
cohort_counts = df.pivot_table(
    index='cohort_month',
    columns=df.groupby('customer_id').cumcount(),
    values='customer_id',
    aggfunc='count'
)

# Retention (% der Kunden, die aktiv bleiben)
cohort_retention = cohort_counts.divide(cohort_counts.iloc[:,0], axis=0)

# Heatmap zeichnen
plt.figure(figsize=(12,6))
sns.heatmap(cohort_retention, annot=True, fmt=".0%", cmap="YlGnBu")
plt.title("Cohort Analysis: Customer Retention")
plt.ylabel("Cohort Month")
plt.xlabel("Months Since Signup")
plt.show()
```

```python
# Dimensionen und Fakten laden
df_customers = pd.read_sql("SELECT * FROM dw.dim_customer", engine)
df_contracts = pd.read_sql("SELECT * FROM dw.dim_contract", engine)
df_transactions = pd.read_sql("SELECT * FROM dw.fact_transaction", engine)
df_churn = pd.read_sql("SELECT * FROM dw.fact_churn", engine)
```

```python
# daten zusammenf端hren
```

```python
# Kunden + Vertrag
df = df_customers.merge(df_contracts, on='customer_id', how='left')

# Kunden + Transaktionen (Summe pro Kunde)
df_rev = df_transactions.groupby('customer_id')['revenue'].sum().reset_index()
df = df.merge(df_rev, on='customer_id', how='left')

# Kunden + Churn
df = df.merge(df_churn[['customer_id', 'churn_reason']], on='customer_id', how='left')
```

```python
#tabelle laden 

df_customers = pd.read_sql("SELECT * FROM dw.dim_customer", engine)
df_contracts = pd.read_sql("SELECT * FROM dw.dim_contract", engine)
df_activity = pd.read_sql("SELECT * FROM dw.fact_customer_activity", engine)
df_payments = pd.read_sql("SELECT * FROM dw.fact_payment", engine)
df_churn = pd.read_sql("SELECT * FROM dw.fact_churn", engine)
```

```python
#Features pro Kunde aggregieren
```

```python
#Kundenaktivit辰t
df_activity_agg = (
    df_activity
    .groupby('customer_id')
    .agg(
        avg_logins=('logins', 'mean'),
        avg_support_calls=('support_calls', 'mean')
    )
    .reset_index()
)
```

```python
#zahlungsverhalten
df_payment_agg = (
    df_payments
    .groupby('customer_id')
    .agg(
        avg_payment_delay=('payment_delay_days', 'mean')
    )
    .reset_index()
)
```

```python
#zusammenfuhren
```

```python
df = (
    df_customers
    .merge(df_contracts[['customer_id', 'contract_length', 'subscription_type']], on='customer_id', how='left')
    .merge(df_activity_agg, on='customer_id', how='left')
    .merge(df_payment_agg, on='customer_id', how='left')
    .merge(df_churn[['customer_id']], on='customer_id', how='left')
)
```

```python
#zielvariable erstellen 

df['churn'] = df['customer_id'].isin(df_churn['customer_id']).astype(int)
```

```python
# Zus辰tzliche Features
```

```python
#Tenure (Kundenalter)
df['signup_date'] = pd.to_datetime(df['signup_date'])
today = pd.Timestamp.today()

df['tenure_days'] = (today - df['signup_date']).dt.days
```

```python
# Missing values behandeln
df.fillna({
    'avg_logins': 0,
    'avg_support_calls': 0,
    'avg_payment_delay': 0
}, inplace=True)
```

```python
#Kategorische Variablen encoden
```

```python
df_model = pd.get_dummies(
    df[['age', 'tenure_days', 'avg_logins', 'avg_support_calls',
        'avg_payment_delay', 'contract_length', 'subscription_type', 'region']],
    drop_first=True
)
```

```python
#Train / test
```

```python
from sklearn.model_selection import train_test_split

X = df_model
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
#logistick regression trenieren 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Model bewerten 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

```python
# Modell interpretieren 
feature_importance = pd.Series(
    model.coef_[0],
    index=X.columns
).sort_values(ascending=False)

print(feature_importance)
```
